```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem, Draw
import py3Dmol

df = pd.read_csv("PI1M.csv")

psmiles = df["SMILES"].iloc[0]
print("Original pSMILES:", psmiles)

# Replace * with hydrogen (capping)
smiles_capped = psmiles.replace("*", "[H]")
print("Capped SMILES:", smiles_capped)

mol = Chem.MolFromSmiles(smiles_capped)
mol = Chem.AddHs(mol)

AllChem.EmbedMolecule(mol, AllChem.ETKDG())
AllChem.MMFFOptimizeMolecule(mol)

# 2D preview
Draw.MolToImage(mol)
```

```python
block = Chem.MolToMolBlock(mol)

view = py3Dmol.view(width=400, height=400)
view.addModel(block, "mol")
view.setStyle({"stick":{}})
view.zoomTo()
view.show()
```
